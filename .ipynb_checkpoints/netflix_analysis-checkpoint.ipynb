```python
# DATA CLEANING 

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('data/netflix_titles.csv')

print("Dataset shape:",df.shape)
print("\nColumn Names:\n",df.columns.tolist())
print("\nMissing Values Before Cleaning:")
print(df.isnull().sum())
```

```python
# HANDLE MISSING VALUES

df['director'] = df['director'].fillna('Unknown') 
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')

df.dropna(subset=['date_added','rating','duration'],inplace=True)

print("Missing Values After Cleaning:")
print(df.isnull().sum())
print("\nDataset Shape After Cleaning:",df.shape)
```

```python
# DUPLICATE CHECK

duplicates = df.duplicated().sum()
print("Duplicated Rows:", duplicates)

title_duplicates = df.duplicated(subset=['title','type']).sum()
print("Duplicated Title:", title_duplicates)

print("Final Dataset Shape:", df.shape)
```

```python
 # CLEANING SUMMARY

print("=" * 45)
print("        DATA CLEANING SUMMARY")
print("=" * 45)
print(f"Original Dataset:     8807 rows, 12 columns")
print(f"After Cleaning:       {df.shape[0]} rows, {df.shape[1]} columns")
print(f"Rows Removed:         {8807 - df.shape[0]}")
print(f"Duplicate Rows:       0")
print(f"Missing Values Left:  {df.isnull().sum().sum()}")
print("=" * 45)
print("Cleaning Steps Done:")
print("  - Director/Cast/Country: Filled with 'Unknown'")
print("  - Rows dropped: Missing date, rating, duration")
print("  - No duplicates found")
print("=" * 45)
```
